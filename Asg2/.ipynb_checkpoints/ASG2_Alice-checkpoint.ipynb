````python
# Imports
from nltk.metrics import BigramAssocMeasures as bam
from gensim.models import Word2Vec, KeyedVectors
from scipy import linalg, dot, cluster
from sklearn import manifold, metrics
from nltk.corpus import wordnet as wn
import gensim.downloader as api
from collections import Counter
from nltk.corpus import reuters
import matplotlib.pyplot as plt
from itertools import chain
from math import log
import gensim.models
import pandas as pd
import numpy as np
import sklearn
import string
import scipy
import nltk
````

````python
# MY CODE Q1 

text = reuters.sents() 

#pos-tagging words
postext = [nltk.pos_tag(word, tagset = "universal") for word in text] 


#Filtering 
# I remove stop words and alphanumeric words (keeps purely alphabetic words)
# and words that are only one character as the text has some random one letter tokens
# I store all extracted (not-removed) words in a list
words = []

stopwords = nltk.corpus.stopwords.words('english')

for sentence in postext:
    filtered = []
    #item is: "word-postag"
    for item in sentence:
        # item[0] = word & item[1] = pos-tag
        if ((len(item[0]) > 1 and item[0].isalpha()) and (not item[0] in stopwords)) :
            filtered.append((item[0].lower(), item[1]))
    words.append(filtered)
    


# Lemmatizing text
# I store the lematized text in a list of lemmatized items
# lemmatized is a list items(words) with their pos-tag
mapping = {"VERB" : wn.VERB, "NOUN" : wn.NOUN, "ADJ" : wn.ADJ, "ADV" : wn.ADV}
lemmatized = []

for sentence in words: 
    lemmatized_sentence = []
    for w, p in sentence:
        if p in [".", "X"]: 
            continue
        elif p in mapping.keys():
            lemma = nltk.WordNetLemmatizer().lemmatize(w, pos = mapping[p])
        else:
            lemma = nltk.WordNetLemmatizer().lemmatize(w)
        lemmatized_sentence.append("-".join([lemma, p]))
        
    lemmatized.append(lemmatized_sentence) 
    
print("lemmatized:",lemmatized[:1])
    
# Counting 
# counting the surface co-occurrences for NOUNS in a symmetric ±5 word collocations span
# this count is found by counting how many times a contextual word occurs
# in a collocational span surrounding the target word

spansize = 5
cooc_surface = Counter()

for sentence in lemmatized: 
    for i,w in enumerate(sentence):
        if w.split("-")[- 1] == "NOUN": 
            span_range = list(range(max(i - spansize, 0), i))
            span_range.extend(range(i + 1, min(i + spansize + 1, len(sentence))))
            for cw in [sentence[idx] for idx in span_range]:
                cooc_surface[(w, cw)] += 1

print("most common co-occurences:",cooc_surface.most_common(15))
````

## Question 2 (20 points)

Weight the counts in the space you created for the previous question by using the following association measures on both spaces:

1. One **measure of your choice** among those available in the [nltk.BigramAssocMeasures](http://www.nltk.org/howto/metrics.html#association-measures) module.
2. The **Positive Local Mutual Information** measure (as shown in class/lab).

**Possible extra**

3. Also use the **smoothed ppmi measure** proposed by [Levy et al. (2015)](http://www.aclweb.org/anthology/Q15-1016). Recall that the authors proposed to smooth the ppmi by raising the context counts to the power of $\alpha$ (where $\alpha= 0.75$ is reported to work well). That is, if $V_c$ is the vocabulary of all the contexts in a given space and $f(c)$ is the context frequency, they proposed the following association measure:

$$PPMI_\alpha (w,c) = max \left(0, \ log_2 \left(\frac{p(w,c)}{p(w) \cdot p_\alpha(c)}\right)  \right) $$

**Question 2:**

````python
# List of measures from the nltk.BigramAssocMeasures module
print([asm for asm in dir(nltk.collocations.BigramAssocMeasures) if not asm.startswith("_")])
````

````python
# Different Counters (frequencies)

# counting lemmas (frequency)
# in the lemmatized list
# is a counter (dictionary like) with keys: words and values: frequency
reuters_lemmas_frequencies = Counter(chain(*lemmatized)) 

# counting lemmas frequency
# filtering for: frequency > 10 and pos-tag: noun, verb, adj, adv
# and not one of the words: "be", "have", "do", "would", "will", "could"
# is a counter (dictionary like) with keys: words and values: frequency
filtered_reuters_lemmas_frequencies = Counter()

# counting the number of noun occurences in filtered_reuters_lemmas_frequencies
# is a counter (dictionary like) with keys: words and values: frequency
selected_reuters_noun_frequencies = Counter()


for k,v in reuters_lemmas_frequencies.items():
    pos = k.split("-")[-1]
    if v >= 10 and \
    pos in ["NOUN", "VERB", "ADJ", "ADV"]\
    and k.split("-")[0] not in ["be", "have", "do", "would", "will", "could"]:
        filtered_reuters_lemmas_frequencies[k] = v
        if pos == "NOUN":
            selected_reuters_noun_frequencies[k] = v

# Counting number of filtered noun co-occurrences
filtered_cooccs_surface = Counter()

for k,v in cooc_surface.items():
    if k[0] in selected_reuters_noun_frequencies and k[1] in filtered_reuters_lemmas_frequencies: #1
        filtered_cooccs_surface[k] = v
        
print(filtered_cooccs_surface.most_common(10))


# defining N to be the sum of the surface co-occurence values (frequencies)
N = sum(cooc_surface.values())
````

````python
# Measure of my choice: chi_sq (chi-squared) 
bam_chi_sq_surface = Counter()

for i, x in filtered_cooccs_surface.items(): 

    bam_chi_sq_surface[i] = bam.chi_sq(x, (reuters_lemmas_frequencies[i[0]], reuters_lemmas_frequencies[i[1]]), N)

print("bigram associate measure's chi squared:", bam_chi_sq_surface.most_common(10)) 

# Positive Local Mutual Information using Positive Pointwise Mutual Information

def ppmi(o_11, r_1, c_1, n):
    """
    Positive Pointwise Mutual Information (Church & Hanks, 1990)
    
    PMI is also available in NLTK:
    from nltk.metrics import BigramAssocMeasures
    print BigramAssocMeasures.pmi(8, (15828, 4675), 14307668)
    """
    observed = o_11
    expected = (r_1*c_1)/n 
    res = log(observed/expected,2)
    return max(0, res)

def plmi(o_11, r_1, c_1, n):
    """
    Positive Local Mutual Information, useful for leveraging the 
    low-frequency bias of the PPMI
    """
    res = o_11 * ppmi(o_11, r_1, c_1, n)
    return res

plmis_surface = Counter()

for i, x in filtered_cooccs_surface.items():

    plmis_surface[i] = plmi(x, reuters_lemmas_frequencies[i[0]], reuters_lemmas_frequencies[i[1]], N)
    
print("plmi:", plmis_surface.most_common(10))

# Smooth PPMI
def smooth_ppmi(o_11, r_1, c_1, n):
    observed = o_11
    probofc = c_1/n
    alph = 0.75
    palphac = (probofc**alph)/(probofc**alph+(1-probofc)**alph)
    
    expected = (r_1/n) * palphac
    res = log(observed/expected,2)
    return max(0, res)

smoothppmi_surface = Counter()

for i, x in filtered_cooccs_surface.items():

    smoothppmi_surface[i] = smooth_ppmi(x, reuters_lemmas_frequencies[i[0]], reuters_lemmas_frequencies[i[1]], N)
    
print("smooth ppmi:", smoothppmi_surface.most_common(10))
````

**Question 3**

Finding the right number of dimensions to keep can be done by looking out how much variance each dimension can account for. I calculate the relative contribution for every single value by by dividing the square of the singular value  𝑘 by the sum of the squares of all the singular values.

Sources: Class Notebooks

````python
# Creating vectors by populating matrix for each distributional
# space made by the differnet types of weighted co-occurrences

# lemma to row/column index mappings
# list of nouns w. pos-tag in alphabetical order
sorted_vectors = sorted(selected_reuters_noun_frequencies)

#dictionary of key:nouns w. pos-tag in alphabetical order value:index
vectors_indices = dict((v,i) for i,v in enumerate(sorted_vectors))

#dictionary of key:filtered words w. pos-tag in alphabetical order value:index
contexts_indices = dict((v,i) for i,v in enumerate(sorted(filtered_reuters_lemmas_frequencies)))

# Distributional Space 1: Chi_sq
# initialize matrix
bamchisqMat = np.zeros((len(vectors_indices), len(contexts_indices)))

# populate the matrix
for pair, weight in bam_chi_sq_surface.items():
    bamchisqMat[vectors_indices[pair[0]]][contexts_indices[pair[1]]] = weight

# Distributional Space 2: PLMI
# initialize matrix
plmiMat = np.zeros((len(vectors_indices), len(contexts_indices)))

# populate the matrix
for pair, weight in plmis_surface.items():
    plmiMat[vectors_indices[pair[0]]][contexts_indices[pair[1]]] = weight

# Distributional Space 3: Smooth PPMI
# initialize matrix
smoothppmiMat = np.zeros((len(vectors_indices), len(contexts_indices)))

# populate the matrix
for pair, weight in smoothppmi_surface.items():
    smoothppmiMat[vectors_indices[pair[0]]][contexts_indices[pair[1]]] = weight
    
dimensions = 100

# The ARPACK wrapper in with the SVD module from SciPy is slower but more efficient than the SVD solver. 
# It reduces more than 8,000 dimensions into 100

bam_chi_sqMat_tr100 = sklearn.decomposition.TruncatedSVD(n_components = dimensions, algorithm = "arpack").fit_transform(bamchisqMat)
plmiMat_tr100 = sklearn.decomposition.TruncatedSVD(n_components = dimensions, algorithm = "arpack").fit_transform(plmiMat)
smoothppmiMat_tr100 = sklearn.decomposition.TruncatedSVD(n_components = dimensions, algorithm = "arpack").fit_transform(smoothppmiMat)
````

````python
# Scipy Modules & Plotting
# decomposing matrix (here we can ignore Vt)
U, s, Vt = linalg.svd(plmiMat)  

rel_variance = ( s**2 / sum(s**2) ) * 100
y_values = rel_variance
x_values = np.arange(len(y_values))

plt.plot(x_values, y_values, linestyle = '-')

plt.xlabel("Singular Values", labelpad = 20)
plt.ylabel("Relative Variance (%)", labelpad = 10)

plt.show()
````
